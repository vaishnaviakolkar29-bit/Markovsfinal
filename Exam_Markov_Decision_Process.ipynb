```python
#import all required libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define states of Markov Process
states = ["Working", "Degraded", "Failed"]
```

```python
# Define Transition probability matrix
P = np.array([
    [0.7, 0.2, 0.1],   # From Working
    [0.3, 0.4, 0.3],   # From Degraded
    [0.6, 0.0, 0.4]    # From Failed
])
```

```python
# Initial state (start with Working)
current_state = 0
```

```python
# Number of steps (time periods)
n_steps = 10
```

```python
# Store state history
state_history = [current_state]
state_history
```

```python
# Perform the Simulation
for step in range(n_steps):
    next_state = np.random.choice([0, 1, 2], p=P[current_state])
    state_history.append(next_state)
    current_state = next_state
    print(current_state)
```

```python
# Convert state indices to names
state_names = [states[s] for s in state_history]
state_names
```

```python
# Plot the state evolution
plt.figure()
plt.plot(state_history, marker='o')
plt.yticks([0, 1, 2], states)
plt.xlabel("Time Step")
plt.ylabel("Machine State")
plt.title("Markov Process Simulation of Manufacturing Machine")
plt.grid()
plt.show()
```

```python
# Count occurrences of each state i.e. frequency of each state
unique, counts = np.unique(state_history, return_counts=True)
print("\nState distribution:")
for u, c in zip(unique, counts):
    print(f"{states[u]}: {c} times")
```

```python
# Import libraries
import networkx as nx
import matplotlib.pyplot as plt

# Create graph
G = nx.DiGraph()

# Add connections
G.add_edge("Working", "Degraded")
G.add_edge("Degraded", "Failed")
G.add_edge("Degraded", "Working")
G.add_edge("Working", "Working")

# Draw graph
nx.draw(G, with_labels=True, node_color='lightblue',
        node_size=3000, font_size=10)

# Show graph
plt.title("Markov Chain Diagram")
plt.show()
```
